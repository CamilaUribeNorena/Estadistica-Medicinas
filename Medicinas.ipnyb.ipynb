```python
# Cargar y mostrar las primeras filas

import pandas as pd
import numpy as np

database = pd.read_csv(r'C:\Users\CamiUribe\OneDrive\Desktop\Medicine_Details.csv')  
print(database.shape)       # filas y columnas
print(database.columns)     # nombres de columnas
print(database.head(5))     # primeras 5 filas
print(database)
```

```python
# hallazgo de nulos y tipos de datos
print(database.dtypes)

#Conteo de nulos
nulos=database.isnull().sum()
print(nulos)
```

```python
# estadistica descriptiva
stats_num = database.describe(include='all')
print(stats_num)
```

```python
# Filtrados
import seaborn as sns
import matplotlib.pyplot as plt

mask = (
    (database["Side_effects"] == 
     "Application site reactions burning irritation itching and redness")
    & 
    (database["Excellent Review %"] >= 80)
)

# Aplico la máscara al DataFrame
maseventossecundarios = database[mask]

# Muestro el resultado
print(maseventossecundarios)

# 1. Calcular frecuencias
count = maseventossecundarios["Medicine Name"].value_counts()
print(count)

# 2. Graficar con barplot usando los arrays de conteo
plt.figure()
sns.barplot(x=count.index, y=count.values)
plt.xticks(rotation=45)
plt.xlabel("Medicine Name")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
